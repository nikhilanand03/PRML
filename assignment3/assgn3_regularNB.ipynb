```python
def NaiveBayes(alpha,trData,trLabels):
    
    num_spam = 0
    for i in range(0,len(trLabels)):
        if(trLabels[i]==1):
            num_spam+=1
    p_cap = num_spam/len(trLabels)

    p = np.zeros((len(trData[0]),2))
    indic_sum = np.zeros((len(trData[0]),2))

    for j in range(len(trData[0])):
        for i in range(len(trData)):
            if(trData[i][j]!=0)and(trLabels[i]==0):
                indic_sum[j][0]+=1
            if(trData[i][j]!=0)and(trLabels[i]==1):
                indic_sum[j][1]+=1

    for k in range(2):
        for j in range(len(trData[0])):
            if(k==0):
                p[j][k] = (indic_sum[j][k]+alpha)/(len(trData[0])*alpha + len(trLabels)-num_spam)
            if(k==1):
                p[j][k] = (indic_sum[j][k]+alpha)/(len(trData[0])*alpha + num_spam)
    
    return p_cap,p,num_spam
```

```python
def convertTxtToArr(st):
    x = st.lower()
    x = x.replace(",", "")
    x = x.replace(":", "")
    x = x.replace(";", "")
    x = x.replace("(", "")
    x = x.replace(")", "")
    x = x.replace("\"", "")
    x = x.replace(".", "")
    x = x.replace("\\", " ")
    x = x.replace("/", " ")
    x = x.replace("'", "")
    x = x.replace("\n"," ")
    x = x.replace("-","")
    arr = x.split()
    return arr
```
